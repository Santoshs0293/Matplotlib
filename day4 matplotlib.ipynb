```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
```

```python
x
```

```python
y = 3*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(x,y,color='g',linestyle=':',linewidth=5)
plt.title('x vs y')
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.show()
```

```python
y1=2*x
y2=3*x
print(y1)
print(y2)
```

```python
plt.plot(x,y1,color='r',linestyle=':',linewidth=3)
plt.plot(x,y2,color='y',linewidth=4)
plt.title('Two lines in a plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color='r',linestyle=':',linewidth=3)

plt.subplot(2,1,2)
plt.plot(x,y2,color='y',linewidth=4)
plt.title('Two lines in a plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```

```python
students = {'bob':87,'julia':45,'anne':100,'matt':67}
```

```python
names=list(students.keys())
```

```python
names
```

```python
values=list(students.values())
values
```

```python
plt.bar(names,values)
plt.show()
```

```python
plt.bar(names,values)
plt.title('Distribution of student marks')
plt.xlabel('Names of students')
plt.ylabel('Marks of students')
plt.show()
```

```python
plt.barh(names,values,color='g')
plt.title('Distribution of student marks')
plt.ylabel('Names of students')
plt.xlabel('Marks of students')
plt.show()
```

```python
x=[1,2,9,4,5,6]
y=[2,5,4,7,6,8]
z=[4,6,3,7,8,9]
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y,marker='*',c='r',s=100)
plt.subplot(1,2,2)
plt.scatter(x,z,c='g',s=200)
plt.show()
```

```python
fruit = ['apple','mango','guava','orange','pineapple']
quantity = [10,5,15,20,8]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['yellow','blue','green','red','orange'])
plt.show()
```

```python
plt.pie(quantity,labels=fruit,radius=2)
plt.pie([100],radius=1,colors=['w'])
plt.show()
```
